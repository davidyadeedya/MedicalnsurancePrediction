```python
#Creating Training and Testing datasets
#Separating Target variable 
X=insurance_df.drop(columns=['charges'])
```

```python
y=insurance_df['charges']
```

```python
y
```

```python
#Converting into Array
X=np.array(X).astype('float32')
y=np.array(y).astype('float32')
```

```python
X
```

```python
y
```

```python
y=y.reshape(-1,1)
```

```python
y.shape
```

```python
#Scaling the data before feeding it to the model
from sklearn.preprocessing import StandardScaler,MinMaxScaler


scaler_x=StandardScaler()
X=scaler_x.fit_transform(X)
scaler_y=StandardScaler()
y=scaler_y.fit_transform(y)
```

```python
#Splitting the dataset into training and Testing Dataset
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
#Use Linear Regression Model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,accuracy_score
```

```python
#Instantinate Model and Train Model
linear_regression_model=LinearRegression()
linear_regression_model.fit(X_train,y_train)
```

```python
#Checking the model Accuracy
linear_model_accuracy=linear_regression_model.score(X_test,y_test)
linear_model_accuracy
```

```python
y_predict=linear_regression_model.predict(X_test)
```

```python
y_predict_orig=scaler_y.inverse_transform(y_predict)
y_test_orig=scaler_y.inverse_transform(y_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
from math import sqrt
k=X_test.shape[1]
n=len(X_test)
RMSE=float(format(np.sqrt(mean_squared_error(y_test_orig,y_predict_orig)),'.3f'))
MSE=mean_squared_error(y_test_orig,y_predict_orig)
MAE=mean_absolute_error(y_test_orig,y_predict_orig)
r2=r2_score(y_test_orig,y_predict_orig)
adj_r2=1-(1-r2)*(n-1)/(n-k-1)
print('RMSE=',RMSE,'\nMSE=',MSE,'\nMAE=',MAE,'\nR2=',r2,'\nAdjR2=',adj_r2)
```
